```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')

# Ganti dengan nama file CSV kamu (File 1)
df = pd.read_csv('dfTransjakarta.csv')  # ganti sesuai nama file
df.head()
```

```python
# Cek info awal
df.info()

# Ubah kolom tapInTime ke datetime
df['tapInTime'] = pd.to_datetime(df['tapInTime'])

# Tambahkan kolom hari dan jam
df['date'] = df['tapInTime'].dt.date
df['hour'] = df['tapInTime'].dt.hour
df['day_name'] = df['tapInTime'].dt.day_name()

df[['tapInTime', 'date', 'hour', 'day_name']].head()
```

```python
daily = df.groupby('date').size()

plt.figure(figsize=(14,5))
daily.plot()
plt.title("Jumlah Penumpang Transjakarta per Hari")
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Penumpang")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
hourly = df.groupby('hour').size()

plt.figure(figsize=(10,5))
hourly.plot(kind='bar', color='skyblue')
plt.title("Distribusi Penumpang per Jam")
plt.xlabel("Jam")
plt.ylabel("Jumlah Penumpang")
plt.xticks(rotation=0)
plt.grid()
plt.show()
```

```python
# Hitung umur dari birth year
df['age'] = 2023 - df['payCardBirthDate']

# Kelompok umur
def label_age(age):
    if age < 25:
        return 'Remaja'
    elif age < 40:
        return 'Dewasa Muda'
    elif age < 60:
        return 'Dewasa'
    else:
        return 'Lansia'

df['age_group'] = df['age'].apply(label_age)

# Visualisasi
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='age_group', palette='Set2')
plt.title("Distribusi Penumpang Berdasarkan Kelompok Umur")
plt.xlabel("Kelompok Umur")
plt.ylabel("Jumlah Penumpang")
plt.grid()
plt.show()
```

```python
sns.countplot(data=df, x='payCardSex', palette='Set1')
plt.title("Distribusi Penumpang Berdasarkan Gender")
plt.xlabel("Gender")
plt.ylabel("Jumlah")
plt.grid()
plt.show()
```

```python
df['payCardBank'].value_counts()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='payCardBank', order=df['payCardBank'].value_counts().index, palette='pastel')
plt.title("Distribusi Penggunaan Kartu Bank oleh Penumpang")
plt.xlabel("Jenis Bank")
plt.ylabel("Jumlah Pengguna")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
top_corridors = df['corridorName'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_corridors.values, y=top_corridors.index, palette='Set3')
plt.title("10 Koridor Transjakarta yang Paling Sering Digunakan")
plt.xlabel("Jumlah Pengguna")
plt.ylabel("Nama Koridor")
plt.grid()
plt.show()
```
